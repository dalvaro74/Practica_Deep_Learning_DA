## Explicacion ficheros:
- **utilidades.py**: Contiene diversas funciones dise√±adas por mi que se utilizaran en varios notebooks
- **models.py**: Fichero python con las funciones de creacion de las redes nn y cnn. Esta extraido del ejemplo indicado por el profesor (https://www.pyimagesearch.com/2019/02/04/keras-multiple-inputs-and-mixed-data/) y modificado para adaptarse a mis necesidades
- **descarga_imagenes.ipynb**: Se utiliza para descargar las imagenes asociadas a cada registro del dataset de airbnb. Ademas de descargar las imagenes en un directorios de drive, genera un nuevo dataset (airbnb-images-clean.csv) solo con aquellos registros que contienen imagenes validas y que seran usados tanto en la red convolucional (cnn_regresion.ipynb) como en la red mixed (mixed_regresion.ipynb).
- **nn_regresion_01**: Utiliza el dataset inicial de airbnb (airbnb-listing.csv) y lleva a cabo sobre el el mismo proceso de limpieza y extraccion de caracteristicas que se llevo a cabo en la practica de machine Learning. Se genera una red neuronal tadicional (regresion) para predecir el precio de las casas a partir de esa informacion.
- **nn_regresion_02.ipynb**: Muy similar al anterior con dos diferecnias. Utiliza el dataset generado en descarga_imagenes.ipynb  (airbnb-images-clean.csv) y genera la red neuronal tradicional de regresion desde la funcion create_nn del fichero models.py, el cual ha sido modificado para la ocasion.
- **cnn_regresion.ipynb**: Vuelve a utillizar el dataset airbnb-images-clean.csv. Con el generara un array con todas las imagenes del dataset con el que alimentara una red convolucional de regresion para predecir el precio de los apartamentos viendo una foto. Como en el caso anterior, la CNN se genera desde la funcion create_cnn del fichero models.py
- **mixed_regresion.ipynb**: En este notebook se generara una red Neuronal que admita tanto datos como imagenes para poder predecir el precio de manera mas eficiente que si se hace por separado desde una cnn solo con imagenes o desde una nn solo con datos.


# Resumen resultados:
Usando como estimador de la bondad del modelo el "mean_absolute_percentaje_error" los resultados de los distintos modelos son los siguientes (para 100 epocas):
- Red Neuronal tradicional (nn_regresion_02) con datos numericos/categoricos del dataset de airbnb: 25,72%
- Red Convolucional (cnn_regresion) con imagenes extraidas del dataset de airbnb: 49,07%
- Red Mixta con datos tanto numericos/categoricos como imagenes: 24,17%

Es necesario hacer notar que si bien los dos primeros modelos convergian rapidamente y a partir de la epoca 60-70 a penas habia mejoria, la red mixta apuntaba a que con mas epocas se hubiera conseguido un resultado aun mejor.
